## 1. Ï†ÑÏ≤¥ ÌïôÏäµÏΩîÎìú

```python
import tensorflow as tf
import numpy as np

learning_rate = 0.01
training_cnt = 10000
display_step = 1000

train_X = np.array([[2, 3], [4, 3], [4, 2], [2, 5], [6, 3], [5, 4] ])
train_Y = np.array([[0], [0], [0], [1], [1], [1]])

X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([2, 1]), name='weight') 
b = tf.Variable(tf.random_normal([1]), name='bias')  

pred = tf.sigmoid(tf.matmul(X ,W) + b)

cost = -tf.reduce_mean(Y * tf.log(pred) + (1 - Y) * tf.log(1 - pred) )
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
op_train = optimizer.minimize(cost)

predicted = tf.cast(pred > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

for epoch in range(training_cnt):
    r_cost, r_W, r_b, r_pred, r_c, r_a, _ = sess.run([cost, W, b, pred, predicted, accuracy, op_train], 
                                                     feed_dict = {X: train_X, Y: train_Y})
    
    if (epoch+1) % display_step == 0:
        print("\nRun_count : [%04d], Train_cost =[%.4f], W =[%.4f %.4f], b =[%.4f], \
               \npred =[%.4f %.4f %.4f %.4f %.4f %.4f], \
               \npred_Y = [%d %d %d %d %d %d ], \
               \ntrue_Y=  [%d %d %d %d %d %d ], \
               \naccuracy = [%.2f%%] " 
              % (epoch+1, r_cost, r_W[0], r_W[1], r_b, 
                 r_pred[0], r_pred[1], r_pred[2], r_pred[3], r_pred[4], r_pred[5],
                 r_c[0], r_c[1], r_c[2], r_c[3], r_c[4], r_c[5],
                 train_Y[0], train_Y[1], train_Y[2], train_Y[3], train_Y[4], train_Y[5],
                 r_a*100))
                   
print("\nOptimization Finished!") 
```

```python
import tensorflow as tf
import numpy as np

learning_rate = 0.01
training_cnt = 10000
display_step = 1000

train_X = np.array([[2, 3], [4, 3], [4, 2], [2, 5], [6,3], [5,4] ])
train_Y = np.array([[0], [0], [0], [1],[1], [1]])

X = tf.placeholder("float", shape=[None, 2])
Y = tf.placeholder("float", shape=[None, 1])

W = tf.Variable(tf.random_normal([2, 1]), name='weight') # input = 2 , output = 1
b = tf.Variable(tf.random_normal([1]), name='bias')  # output = 1

pred = tf.sigmoid(tf.matmul(X ,W) + b)

cost = -tf.reduce_mean(Y * tf.log(pred) + (1 - Y) * tf.log(1 - pred) )
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
op_train = optimizer.minimize(cost)

predicted = tf.cast(pred > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```

### 1) ÌååÎùºÎ©îÌÑ∞ Í∞í ÏÑ§Ï†ï

```python
# ÌååÎùºÎ©îÌÑ∞Í∞í ÏÑ§Ï†ï
learning_rate = 0.01
training_cnt = 10000
display_step = 100  # ÏõêÌïòÎäî Ï∂úÎ†• ÏúÑÏπò Ï°∞Ï†ï
```

### 2) Ìä∏Î†àÏù¥Îãù Îç∞Ïù¥ÌÑ∞ Î≥ÄÏàò ÏÑ†Ïñ∏
- ÏûÖÎ†•ÏúºÎ°ú Îì§Ïñ¥Í∞ÄÎäî x data(input 2Í∞ú), y data(output 1Í∞ú) ÏÑ§Ï†ï
- numpy arrayÎ•º ÏÇ¨Ïö© 

```python
train_X = np.array([[2, 3], [4, 3], [4, 2], [2, 5], [6, 3], [5, 4] ])
train_Y = np.array([[0], [0], [0], [1], [1], [1]])
```

### 3) tf graph input
- X : Îì§Ïñ¥Ïò§Îäî rowÎäî Ï†ïÌï¥ÏßÑÍ≤å ÏóÜÍ≥†, column ÏùÄ 2Í∞ú Ï¶â ÏûÖÎ†•Î≥ÄÏàò 2Í∞ú 
- Y : Îì§Ïñ¥Ïò§Îäî rowÎäî Ï†ïÌï¥ÏßÑÍ≤å ÏóÜÍ≥†, column ÏùÄ 1Í∞ú Ï¶â output 1Í∞ú
- matrix ÏÇ¨Ïö© ÌïòÍ∏∞ ÎïåÎ¨∏Ïóê ÏûÖÎ†• Î≥ÄÏàòÎ•º Îã¥Îäî placeholder 1Í∞úÎ°ú ÎêúÎã§

```python
X = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

### 4) tf.random_normal
- bias, weight Ï¥àÍ∏∞Í∞íÏùÑ ÎÇúÏàòÎ°ú ÏÉùÏÑ±

```python
W = tf.Variable(tf.random_normal([2, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

### 5) sigmoid Ìï®Ïàò ÏÇ¨Ïö©
- Í∏∞Ï°¥ pred Í≥ÑÏÇ∞Ïóê sigmoid Ìï®ÏàòÎ•º Ï†ÅÏö©Ìï¥ 0~1ÏÇ¨Ïù¥Ïùò Í∞íÏúºÎ°ú Î≥ÄÌôò
- sigmoid function => H(X) = 1/(1+ùëí^‚àíz )
- z = XW + b


```python
pred = tf.sigmoid(tf.matmul(X ,W) + b)
```

### 6) cost function
- 0~1ÏÇ¨Ïù¥Ïùò Í∞íÏùÑ Í∑ºÏÇ¨Ìôî ÌïòÍ∏∞ ÏúÑÌï¥ÏÑú logÌï®ÏàòÎ•º ÏÇ¨Ïö©
- ùê∂(ùêª(ùë•),ùë¶)=1/ùëö ‚àë(ùë¶ùëôùëúùëî(ùêª(ùë•)) ‚àí(1 ‚àíùë¶)log‚Å°(1 ‚àíùêª(ùë•)) 

```python
cost = -tf.reduce_mean(Y * tf.log(pred) + (1 - Y) * tf.log(1 - pred) )
```

### 7) ÌïôÏäµ
- Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ï ÏÇ¨Ïö©

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate)
op_train = optimizer.minimize(cost)
```

### 8) ÌïôÏäµÎêú ÏòàÏ∏°Í∞íÏùÑ 0Í≥º 1Î°ú Î≥ÄÌôò
- 0~1ÏÇ¨Ïù¥Î°ú ÌïôÏäµÎêú ÏòàÏ∏°Í∞íÏùÑ 0Í≥º 1Î°ú ÎÇòÎàÑÏñ¥ Î∂ÑÎ•òÌï† Ïàò ÏûàÎèÑÎ°ù ÎßåÎì¨

```python
predicted = tf.cast(pred > 0.5, dtype=tf.float32)
```

### 9) Ï†ïÌôïÎèÑ
- accuracyÎ•º Í≥ÑÏÇ∞ÌïòÏó¨ Î∂ÑÎ•òÍ∞Ä Ï†ïÌôïÌïúÏßÄ ÌôïÏù∏
- ÏòàÏ∏°Í∞íÍ≥º Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞Ïùò ÏùºÏπò Ïó¨Î∂Ä Í≥ÑÏÇ∞ 
- ÏïÑÎûò ÏΩîÎìúÎäî ÌèâÍ∑†ÏùÑ Ïù¥Ïö©Ìïú Ï†ïÌôïÎèÑ Í≥ÑÏÇ∞

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```
